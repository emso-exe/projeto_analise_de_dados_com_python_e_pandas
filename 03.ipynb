```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
df = pd.read_excel("datasets/AdventureWorks.xlsx")
```

```python
df["Valor Venda"].sum()
```

```python
df["custo"] = df["Custo Unitário"].mul(df["Quantidade"])
```

```python
round(df["custo"].sum(), 2)
```

```python
df["lucro"] = df["Valor Venda"] - df["custo"]
df.head()
```

```python
round(df["lucro"].sum(), 2)
```

```python
df["Tempo_envio"] = (df["Data Envio"]-df["Data Venda"]).dt.days
df.head()
```

```python
df["Tempo_envio"].dtype
```

```python
df.groupby("Marca")["Tempo_envio"].mean()
```

```python
df.isnull().sum()
```

```python
df.groupby([df["Data Venda"].dt.year, "Marca"])["lucro"].sum()
```

```python
pd.options.display.float_format='{:20,.2f}'.format
```

```python
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])[
    "lucro"].sum().reset_index()
lucro_ano
```

```python
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

```python
df.groupby("Produto")["Quantidade"].sum().sort_values(
    ascending=True).plot.barh(title="Total Produtos Vendidos")
plt.xlabel("Total")
plt.ylabel("Produtos")
```

```python
df.groupby(df["Data Venda"].dt.year)[
    "lucro"].sum().plot.bar(title="Lucro X Ano")
plt.xlabel("Ano")
plt.ylabel("Receita")
```

```python
df.groupby(df["Data Venda"].dt.year)["lucro"].sum()
```

```python
df_2009 = df[df["Data Venda"].dt.year == 2009]
df_2009.head()
```

```python
df_2009.groupby(df_2009["Data Venda"].dt.month)[
    "lucro"].sum().plot(title="Lucro X Mês")
plt.xlabel("Mês")
plt.ylabel("Lucro")
```

```python
df_2009.groupby("Marca")["lucro"].sum().plot.bar(title="Lucro X Marca")
plt.xlabel("Marca")
plt.ylabel("Lucro")
plt.xticks(rotation='horizontal')
```

```python
df_2009.groupby("Classe")["lucro"].sum().plot.bar(title="Lucro X Classe")
plt.xlabel("Classe")
plt.ylabel("Lucro")
plt.xticks(rotation='horizontal')
```
